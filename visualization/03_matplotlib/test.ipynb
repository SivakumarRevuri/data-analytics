```python
from matplotlib import pyplot as pt
```

```python
Names = ['Arun','James','Ricky','Patrick']
Marks = [51,87,45,67]

pt.title('Result')
pt.xlabel('Names')
pt.ylabel('Marks')
```

```python
pt.scatter(Names,Marks,color = 'blue')
pt.show()
```

```python
pt.plot(Names,Marks,'b*--')
pt.show()
```

```python
pt.bar(Names,Marks,color = 'blue')
pt.show()
```

```python
pt.plot(Names,Marks,'r--')
pt.show()
```
